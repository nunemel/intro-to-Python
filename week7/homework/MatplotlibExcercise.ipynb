```python
from pandas import DataFrame
import numpy as np


#The productivity values for the 3 groups are the following: 

g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
index = ['M','T','W','Th','F']

a = {'Group1': g1, 'Group2': g2, 'Group3' : g3}

df = DataFrame(a, index = index)
print(df)
ax = df.plot(kind = 'bar', title='Productivity during the week', rot = 0)
ax.set_xlabel('Groups')
ax.set_ylabel('Productivity')

```

```python
import numpy as np 
from matplotlib import pyplot as plt
import random

number_of_circles = 25 
coords = [0, 5, 10, 15, 20, 25, 30]
x = [random.randrange(1, 31) for i in range(number_of_circles)] 
y = [random.randrange(1, 31) for i in range(number_of_circles)]
colors = np.array([np.random.randint(256, size = 3) for i in range(number_of_circles)]) 
area = [random.randrange(20, 501) for i in range(number_of_circles)]
plt.xticks(coords)
plt.yticks(coords)
plt.ylim(bottom = 0, top = 30)
plt.xlim(left = 0, right = 30)
plt.xlabel("X")
plt.ylabel("Y")
plt.scatter(x, y, s = area, c = colors/255.0)
plt.title("Colorful dots")
plt.show()
```
